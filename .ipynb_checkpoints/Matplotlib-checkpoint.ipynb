# **Matplotlib**

**What is Matplotlib**

Matplotlib is a powerfull python library used for data visualization<br>
it helps you create wide variety of static,animated and interactive plots easily

```python
import matplotlib.pyplot as plt
```

**Basic plot**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.plot(x,y)
plt.show()
```

Explanation:<br>
* plt.plot()->draws a line graph
* plt.show()->displays the graph

**Adding label and title**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.plot(x,y)

plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

**Customizations**

1. plot customization

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.plot(x, y, color="#8231de",              #line color 
                linestyle="dotted",         #linestyle=solid(default),dashed,dotted,etc
                linewidth=4,
                marker="o",                    #marker=.,o,*,v,etc
                markersize=10,                 #ms
                markerfacecolor="#de28eb",      #mfc
                markeredgecolor="#de28eb")      #mec
    

plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
# if you want this customization in more than one plot you can use tuple unpacking
# example:
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y1=np.array([2,4,6,8,10])
y2=np.array([3,6,9,12,15])
y3=np.array([4,8,12,16,20])

line_style=dict(color="#8231de",linestyle="dotted",     
                linewidth=4,
                marker="o",                    
                markersize=10,                 
                markerfacecolor="#de28eb",      
                markeredgecolor="#de28eb")     

plt.plot(x, y1, **line_style)           #(**= tuple unpacking)
plt.plot(x, y2, **line_style)    
plt.plot(x, y3, **line_style)

plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

2. Label and title customizations

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

plt.plot(x, y, color="#8231de",              #line color 
                linestyle="dotted",         #linestyle=solid(default),dashed,dotted,etc
                linewidth=4,
                marker="o",                    #marker=.,o,*,v,etc
                markersize=10,                 #ms
                markerfacecolor="#de28eb",      #mfc
                markeredgecolor="#de28eb")      #mec
    

plt.title("Simple Line Plot", color="#873378",
                              fontsize=20,
                              fontweight="bold",
                              family="Arial"
         )

plt.xlabel("X-Axis ", color="#fc00cf",
                      fontsize=10,
                      fontweight="bold",
                      family="Arial")
plt.ylabel("Y-Axis ",  color="#fc00cf",
                      fontsize=10,
                      fontweight="bold",
                      family="Arial")

# customizing ticks
plt.xticks(x)
plt.tick_params(axis="both",colors="#f274dc")

plt.show()
```

**Bar Charts**

```python
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 12]

plt.bar(x, y, color='orange')
plt.title("Bar Chart Example")
plt.show()
```

Horizontal barchart:

```python
plt.barh(x, y, color='skyblue')
plt.show()
```

**Scatter plot**

```python
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 12]
plt.scatter(x, y, color='red')
plt.title("Scatter Plot Example")
plt.show()
```

**Pie Charts**

```python
sizes = [30, 20, 25, 25]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

```python
# explode->explode parameter is used to visually separate one or more wedges (slices) from the center of the pie. 

sizes = [30, 20, 25, 25]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=[0,0,0,0.1])
plt.title("Pie Chart Example")
plt.show()
```

**Histogram**

```python
data = [10, 20, 20, 30, 40, 40, 50, 60, 60, 70, 80]
plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.show()
```
